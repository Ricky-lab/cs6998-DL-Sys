```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # For reproducibility
x_values = np.linspace(0, 10, 20)  # Generate 20 points between 0 and 10
noise = np.random.normal(0, np.sqrt(0.3), size=x_values.shape)  # N(0, 0.3)
f_x_values = x_values + np.sin(1.5 * x_values)  # f(x) = x + sin(1.5x)
y_values = f_x_values + noise  # y(x) = f(x) + noise

# Create a scatter plot for y(x) (noisy values)
plt.scatter(x_values, y_values, color='blue', label='Noisy y(x)')

# Plot smooth line for f(x) (true values without noise)
x_smooth = np.linspace(0, 10, 1000)  # More points for a smooth curve
f_x_smooth = x_smooth + np.sin(1.5 * x_smooth)
plt.plot(x_smooth, f_x_smooth, color='red', label='True f(x)', linewidth=2)

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of y(x) and True Function f(x)')
plt.legend()
plt.grid(True)

plt.show()
```
